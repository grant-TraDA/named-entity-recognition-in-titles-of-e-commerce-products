```python
import os
import pandas as pd
import json
import re


PL_WDC_RAW_DATA_PATH = os.path.join(
    os.path.abspath(''),
    '../raw_data.csv'
)

data = pd.read_csv(PL_WDC_RAW_DATA_PATH,
                   sep=',',
                   index_col=False )
data['title_original'] = data['title']
data['title'] = data['title'].str.lower().astype(str)
data['brand'] = data['brand'].str.lower().astype(str)
data['category'] = data['category'].str.lower().astype(str)
print(data.head(n=10))
```

```python
import spacy


def find_brand(product, data_tags):
    data_sentence = str(product['title'])
    brand = product['brand']
    pos = data_sentence.find(brand)
    n_brand = brand.count(' ') + 1
    if pos > -1:
        first_position = data_sentence[:pos].count(' ')
        for i_brand in range(n_brand):
            tag = 'B-PER'
            if i_brand > 0:
                tag = 'I-PER'
            data_tags[first_position+i_brand] = tag


def find_units(product, data_tags):
    title = product['title']
    p = re.compile(r'\d+ [a-z]+')
    for m in p.finditer(title):
        first_position = title[:m.start()].count(' ')
        data_tags[first_position] = 'B-ORG'
        data_tags[first_position+1] = 'I-ORG'


nlp = spacy.load('pl_core_news_md')

def find_kind(product, data_tags):
    data_sentence = product['title'].split()
    categories = product['category'].split('>')[:-1]
    categories = [category.strip() for category in categories]
    category_set = set()
    for category in categories:
        category_parts = category.lower().split()
        for part in category_parts:
            if part in nlp.Defaults.stop_words:
                continue
            doc = nlp(part.lower())
            category_set.add(doc[0].lemma_)
    for idx, word in enumerate(data_sentence):
        doc = nlp(word.lower())
        if doc[0].lemma_ in category_set:
            if idx > 0 and data_tags[idx-1] in ['B-MISC', 'I-MISC']:
                data_tags[idx] = 'I-MISC'
            else:
                data_tags[idx] = 'B-MISC'
```

```python
import re

ner_data_sentences = []
ner_data_tags = []

data_dict = data.to_dict('records')

print(data_dict)
ids = []
for idx, product in enumerate(data_dict):
    print(f'{idx}/{len(data_dict)}')

    try:
        product_data = product['title'].split()
        product_tags = ['O'] * len(product_data)
        find_brand(product, product_tags)
        find_units(product, product_tags)
        find_kind(product, product_tags)
        ner_data_sentences.append(product['title_original'].split())
        ner_data_tags.append(product_tags)
        ids.append(idx)
    except:
        pass
    # print(product)
    # print(product_tags)
    # print(product['title_original'].split())
    # print("# # # # # #")



for sentence, tags in zip(ner_data_sentences, ner_data_tags):
    print(sentence)
    print(tags)
    print("# # # # #")
```

```python
output_count = len(ner_data_sentences)

RESULTS_PATH = os.path.join(os.path.abspath(''), "../ner_dataset3")

parts = [
    {"label": "train", "start": 0.0, "end": 0.8},
    {"label": "test", "start": 0.8, "end": 1.0},
    {"label": "valid", "start": 0.8, "end": 1.0},
    {"label": "full", "start": 0.0, "end": 1.0}
]

for part in parts:
    part_path = os.path.join(RESULTS_PATH, f"{part['label']}.txt")
    start = int(output_count * part['start'])
    end = int(output_count * part['end'])

    result_file = open(part_path, 'w+')

    for idx in range(start, end):
        data_sentence = ner_data_sentences[idx]
        data_tags = ner_data_tags[idx]
        for token, tag in zip(data_sentence, data_tags):
            result_file.write(f'{token}\t{tag}\n')
        result_file.write('\n')
    result_file.close()
```
